```python
import glob
files = glob.glob('../data/Videos/UDIVA/**/**/FC1/*.pt')
```

```python
df = pd.read_csv("data/train.csv")
speaker_paths = df["speaker_path"]
speaker_paths
```

```python
file_names = list(filter(lambda x: 'marlin' not in x, files))
```

```python
from dtaidistance import dtw_ndim
from tqdm import tqdm

distances = []
for fname in tqdm(file_names):
    # Load the .pt file
    file_path = f'{fname}'
    arr = torch.load(file_path)
    series = arr.detach().cpu().numpy()

    # Calculate the distance between the series
    distances.append(series)
```

```python
distances
```

```python
matrix = [[0] * len(distances) for _ in range(len(distances))]
```

```python
for i in tqdm(range(len(distances))):
    for j in range(i + 1, len(distances)):
        series1 = distances[i]
        series2 = distances[j]
        d = dtw_ndim.distance(series1, series2)
        matrix[i][j] = d
        matrix[j][i] = d
```

```python
for i in tqdm(range(len(matrix))):
    for j in range(len(matrix[i])):
        print(f"Distance between {file_names[i]} and {file_names[j]}: {matrix[i][j]}")
```

```python
def get_similar_ins(ins):
    common_indices = []
    if ins.ndim == 1:
        for j in range(ins.shape[0]):
            if ins[j] == 1:
                common_indices.append((j,))
    else:
        for i in range(ins.shape[0]):
            for j in range(ins.shape[1]):
                if ins[i, j] == 1:
                    common_indices.append((i, j))
    return common_indices
```
